## What is nifti

NIfTI images is a file format supported by most neuroimaging software packages

### Load packages

These will be needed to perform analysis

```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import nibabel as nib
```

### Load file

Load the rest image from the current directory

```python
fmri_file = 'rest.nii.gz'
img = nib.load(fmri_file)
```

Determine the type of the image

```python
print('Image datatype:',type(img))
```

Determine the dimesions of the image

```python
print('Image shape:', img.shape)
```

## Three parts of nifti images

Images can be divided into ~ three parts:

1. header with metadata

2. image data

3. affine matrix

### Header

The header contains metadata about the scan. This includes units of measurement, voxel size, etc.

The header is an attribute

```python
hdr = img.header
```

Get the voxel size with `get_zooms()`

```python
hdr.get_zooms()
```

Get the units of the measurements

The voxel size is measured in milimeters

The time is measured in seconds

```python
hdr.get_xyzt_units()
```

### Data

Loading the image doesn't load the actual image data. This is because the image data takes up a lot of memory

Get the data by using `get_fdata()`. This returns a numpy array with same dimensions as image data

```python
img_data = img.get_fdata()

print('Type of image data:',type(img_data))
print('Shape of image data:',img_data.shape)
```

Look into the data

```python
print('Image data:\n\n', img_data)
```

Image data is just numbers, like any image. Higher number = more signal scanner recorded

Relative difference between voxel values is most important.

This shows low numbers since these are voxels at the edge of the scan.

Take sample from the middle to get more intense values

```python
mid_vox = img_data[30:33, 30:33, 30:33, 0]
print('Middle voxels:\n\n', mid_vox)
```

Exact values aren't interpretable since scale is not standard. But this helps demonstrate the range since these values are much higher than the edge voxels

fMRI scans can be visualized by plotting numbers by assigning colors to numbers based on intensity. Graph a slice of the data since can't plot 3D image

```python
mid_slice_x = img_data[32,:,:,0]

print('Shape of X slice:', mid_slice_x.shape)
```

Use matplotlib to plot this slice using `imshow()`

```python
plt.imshow(mid_slice_x.T, cmap='gray', origin='lower')
plt.xlabel('First axis')
plt.ylabel('Second axis')
plt.colorbar(label='Signal intensity')
plt.show()
```

Look at fMRI data from the time dimension by extracting a single voxel's time series.

First, extract time series for the middle voxel across all spatial dimensions

```python
mid_vox_ts = img_data[32, 32, 17, :]

print('Voxel timeseries shape: %s' %(mid_vox_ts.shape,))
```

Visualize the same point over 20 time points

```python
from matplotlib import patches

fig, axes = plt.subplots(ncols=5, nrows=4, figsize=(20, 10))  # 20 timepoints
# Loop over the first 20 volumes/timepoints
for t, ax in enumerate(axes.flatten()):    
    ax.imshow(img_data[32, :, :, t].T, cmap='gray', origin='lower')  # index with t!
    rect = patches.Rectangle((38, 20), 2, 2, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    ax.axis('off')
    ax.set_title('t = %i' % t, fontsize=20)
fig.tight_layout()
```

It is very difficult to determine if the point has changed intensity in the time series because fluctuations are usually 1-3% compared to baseline (average) activity

```python
plt.figure(figsize=(18,2))
plt.plot(mid_vox_ts, 'o-', ms=4)
plt.xlim(-1, mid_vox_ts.size)
plt.ylim(810, 930)
plt.xlabel('Time (nr. of volumes)', fontsize=15)
plt.ylabel('Signal Intensity', fontsize=15)
plt.show()
```

Blue dots represent signal intensity plotted across time

### Affine

Each file contains an affine matrix

This relates the position of the image coordinates to real world coordinates. 

Only need a single matrix to transform image data into real world coordinates

```python
np.set_printoptions(suppress=True, precision=3)  # suppresses scientific notation
A = img.affine
print(A)
```

```python
import matplotlib.patches as patches
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

ax[0].imshow(img_data[16, :, :, 0].T, origin='lower', cmap='gray')
ax[0].set_xlabel('Second dim voxel coords.', fontsize=12)
ax[0].set_ylabel('Third dim voxel coords', fontsize=12)
ax[0].set_title('First dimension, slice nr. 70', fontsize=15)
rect = patches.Rectangle((20, 16), 3, 3, linewidth=2, edgecolor='r', facecolor='none')
ax[0].add_patch(rect)

ax[1].imshow(img_data[:, 32, :, 0].T, origin='lower', cmap='gray')
ax[1].set_xlabel('First dim voxel coords.', fontsize=12)
ax[1].set_ylabel('Third dim voxel coords', fontsize=12)
ax[1].set_title('Second dimension, slice nr. 100', fontsize=15)
rect = patches.Rectangle((20, 16), 3, 3, linewidth=2, edgecolor='r', facecolor='none')
ax[1].add_patch(rect)

ax[2].imshow(img_data[:, :, 32, 0].T, origin='lower', cmap='gray')
ax[2].set_xlabel('First dim voxel coords.', fontsize=12)
ax[2].set_ylabel('Second dim voxel coords', fontsize=12)
ax[2].set_title('Third dimension, slice nr. 100', fontsize=15)
rect = patches.Rectangle((20, 16), 3, 3, linewidth=2, edgecolor='r', facecolor='none')
ax[2].add_patch(rect)

fig.tight_layout()
```

Get the real world coordinate

```python
xyz1 = A @ np.array([20, 16, 16, 1])
print('Real world coords:', xyz1)
```
