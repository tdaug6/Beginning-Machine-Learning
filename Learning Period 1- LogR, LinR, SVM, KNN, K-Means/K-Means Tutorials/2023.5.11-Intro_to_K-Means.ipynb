# Intro to K-Means on W3School

K-means cluster data together.

This is accomplished by:
1. Choosing a random initial point for k clusters
2. Assigning each point to the closest cluster "center"
3. Calculate a new center
4. Reassign each point to the closest center
5. Repeating until the center points no longer changes

## Making an algorithm

**1.** Import modules

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

**2.** Create arrays that resemble two variables in a dataset

```python
x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]
```

**3.** Combine data to set of points

```python
data = list(zip(x,y))

print(data)
```

**4.** Determine best value for k.

Maximum number of clusters is 10 since we have 10 points

Run K-means for range of possible values (1-10) and determine inertia

```python
# Hold all inertias to plot
inertias = []

# Loop through k values
for i in range(1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(data)
    inertias.append(kmeans.inertia_)

# Plot results
plt.plot(range(1,11),inertias,marker='o')
plt.title("Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

"elbow" at K = 2 (where inertia becomes more linear

**5.** Fit KMeans where k = 2 and plot the results

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

plt.scatter(x,y,c=kmeans.labels_)
plt.show()
```
