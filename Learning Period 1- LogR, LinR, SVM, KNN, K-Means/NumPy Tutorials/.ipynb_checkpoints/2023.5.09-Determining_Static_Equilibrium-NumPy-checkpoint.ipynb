```python
import numpy as np
import matplotlib.pyplot as plt
```


## Vectors using NumPy

**1.** Create vectors for forces

```python
forceA = np.array([1,0,0])
forceB = np.array([0,1,0])
print("Force A = ",forceA)
print("Force B = ",forceB)
```

**2.** Plot vectors 

```python
fig = plt.figure()
d3 = fig.add_subplot(projection="3d")

d3.set_xlim(-1,1)
d3.set_ylim(-1,1)
d3.set_zlim(-1,1)

x,y,z = np.array([0,0,0])

u,v,w = forceA #Break vector into components
d3.quiver(x,y,z,u,v,w, color="r",label="forceA")

u,v,w = forceB #Break vector into components
d3.quiver(x,y,z,u,v,w, color="b",label="forceB")

plt.legend()
plt.xticks([-1,-0.5,0,0.5,1])
plt.yticks([-1,-0.5,0,0.5,1])


plt.show()
```

**3.** Add forces

```python
forceC = forceA + forceB

fig = plt.figure()
d3 = fig.add_subplot(projection="3d")

d3.set_xlim(-1,1)
d3.set_ylim(-1,1)
d3.set_zlim(-1,1)

x,y,z = np.array([0,0,0])

u,v,w = forceA
d3.quiver(x,y,z,u,v,w, color="r",label="forceA")

u,v,w = forceB
d3.quiver(x,y,z,u,v,w, color="b",label="forceB")

u,v,w = forceC
d3.quiver(x,y,z,u,v,w, color="g",label="forceC")

plt.legend()

plt.show()
```


**4.** Create opposing force to achieve static equilibrium

```python
R = np.array([-1,-1,0])

fig = plt.figure()

d3 = fig.add_subplot(projection="3d")

d3.set_xlim(-1,1)
d3.set_ylim(-1,1)
d3.set_zlim(-1,1)

x,y,z = np.array([0,0,0])

u,v,w = forceC
d3.quiver(x,y,z,u,v,w, color="g",label="forceC")

u,v,w = R
d3.quiver(x,y,z,u,v,w, color="y",label="R")

u,v,w = forceC + R
d3.quiver(x,y,z,u,v,w, color="r",label="forceC + R")

plt.legend()
plt.show()
```


## Solve Equilibrium as sum of moments

Used when forces aren't applied at same point

**1.** Solve for reaction force and moment when 5N force is applied perpendicularaly to base of 2m pole

```python
f = 5
L = 2

R = 0 - f
M = 0 - f * L

print("Reaction force = ",R)
print("Reaction moment = ", M)
```


## Find values with physical properties


```python
poleBase = np.array([0,0,0])
cordBase = np.array([3,0,0])
cordConnection = np.array([0,0,2])

poleDirection = cordConnection - poleBase
print("pole direction = ",poleDirection)

cordDirection = cordBase - cordConnection
print("cord direction = ",cordDirection)
```


**2.** Convert to unit vector

```python
cordUnit = cordDirection / np.linalg.norm(cordDirection)
print("Cord unit vector = ",cordUnit)
```


**3.** Multiply with magnitude of force to find force vector

```python
cordTension = 5
forceCord = cordTension * cordUnit
print("Force from cord = ",forceCord)
```


**4.** Find cross product of force vector and radius to find the moment of the cord

```python
cordMoment = np.cross(forceCord,poleDirection)
print("Moment of cord = ",cordMoment)
```


**5.** Find reaction force and moment using previous steps

```python
equilibrium = np.array([0,0,0])

R = equilibrium-forceCord
print("Reaction force = ",R)

M = equilibrium - cordMoment
print("Reaction moment",M)
```
