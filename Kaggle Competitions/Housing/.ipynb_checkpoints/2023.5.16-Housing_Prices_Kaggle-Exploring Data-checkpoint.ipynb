# Kaggle Housing Prices - EDA

This notebook contains some data exploration for the Housing Prices Kaggle competition. 
The purpose of this file is to examine the relationship between features and the target in the training dataset.  

This was written before knowing about correlation matrices.

## Visualize data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')

df.head()
```

```python
y = df['SalePrice']
```

```python
plt.scatter(df['MSSubClass'], y)
plt.xlabel('MSSubClass')
plt.ylabel('Sale Price')

plt.title('Sale Price vs. MSSubClass')

plt.show()
```
